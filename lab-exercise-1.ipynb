# EXERCISE 5:
A python function that takes a positive integer and returns the sum of the cube of all the positive integers smaller than the specified number.

```python
x = int(input("enter a number:"))
put = list(range(1,x))

y = 0
for i in put:
    n = i**3
    y = y + n
print(y)
```
